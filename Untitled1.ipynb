```python
df.dropna(inplace=True)
print("After cleaning:",df.shape)
```

```python
plt.figure(figsize=(12,6))
df['location'].value_counts().head(10).plot(kind='bar',color='orange')
plt.title('Top 10 Locations in Bangalore')
plt.xlabel('Location')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.savefig('/content/drive/MyDrive/chart1_locations,png')
plt.show()
```

```python
plt.figure(figsize=(10,6))
df['online_order'].value_counts().plot(kind='pie',autopct='%1.1f%%',colors=['lightgreen','lightcoral'])
plt.title('Online Order Availability')
plt.savefig('/content/drive/MyDrive/chart2_online_order.png')
plt.show()
```

```python
plt.figure(figsize=(10,6))
df['rest_type'].value_counts().head(10).plot(kind='bar', color='skyblue')
plt.title('Top 10 Restaurant Types')
plt.xlabel('Restaurant Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('/content/drive/MyDrive/chart3_rest_type.png')
plt.show()
```

```python
plt.figure(figsize=(10,6))
df['rate'].value_counts().head(10).plot(kind='bar', color='purple')
plt.title('Restaurant Ratings Distribution')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('/content/drive/MyDrive/chart4_ratings.png')
plt.show()
```
